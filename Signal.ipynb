```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
t_continuous=np.linspace(0,1,500)
```

```python
signal_continuous=np.sin(2*np.pi*5*t_continuous)
```

```python
t_discrete = np.arange(0, 1, 0.1)
```

```python
signal_discrete = np.sin(2 * np.pi * 5 * t_discrete)
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 1)
```

```python
plt.plot(t_continuous, signal_continuous)
```

```python
plt.title('Continuous Time Signal')
```

```python
plt.xlabel('Time')
```

```python
plt.ylabel('Amplitude')
```

```python
plt.subplot(1, 2, 2)
```

```python
plt.stem(t_discrete, signal_discrete, use_line_collection=True)
```

```python
plt.title('Discrete Time Signal')
```

```python
plt.xlabel('Time')
```

```python
plt.ylabel('Amplitude')
```

```python
plt.tight_layout()
```

```python
plt.show()
```
